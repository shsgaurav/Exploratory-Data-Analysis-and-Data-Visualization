```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
iris = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
iris['species'] = pd.DataFrame(data.target)
```

```python
iris.head()
```

```python
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=iris, kind='kde')
```

```python
iris_setosa  = iris[iris['species']==0]
```

```python
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=iris_setosa, kind='kde')
```

```python
 haber = pd.read_csv("haberman.csv", header=None)
```

```python
haber.describe()
```

```python
haber.info()
```

```python
haber.plot(x=0,y=1, kind='scatter')
```

```python
sns.pairplot(haber, hue=3)
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(haber, hue=3, size=4) \
.map(sns.distplot, 2) \
.add_legend()
```

```python
counts, bin_edges = np.histogram(haber[2], bins=10, density=True)
pdf = counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```
