```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel("Numbers")
plt.grid()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')#R-red color o-0dotted
plt.ylabel("Numbers")
plt.grid()
```

```python
import numpy as np
t = np.arange(0,5,0.2)
```

```python
plt.plot(t,t**2, 'b--', label='^2')#blue dashes
plt.plot(t,t**2.2, 'rs', label='^2.2')#red squares
plt.plot(t,t**2.5, 'g^', label='^2.5')#green triangle
plt.grid()
plt.legend()
```

```python
X,y = [1,2,3,4],[1,4,9,16]
```

```python
plt.plot(X,y, linewidth=5)
plt.grid()
```

```python
x1, y1 = [1,2,3,4],[1,4,9,16]
x2, y2 = [5,6,7,8],[25,30,35,40]
```

```python
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[0], color='g', linewidth=2.0)
plt.grid()
```

```python
t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.2)
plt.subplot(211)
plt.plot(t1, np.sin(t1), 'g-')
plt.grid()
plt.subplot(212)
plt.plot(t2, np.cos(t2),'r-')
plt.grid()
```
