```python
import numpy as np
import scipy.stats as stats
import pylab
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
std_normal = np.random.normal(loc = 0, scale=1, size=10000)
```

```python
stats.probplot(std_normal, dist='norm', plot=pylab)
pylab.show()
```

```python
stats.skew(std_normal)
```

```python
stats.kurtosis(std_normal)
```

```python
uniform = np.random.uniform(low=-1, high=1, size=10000)
stats.probplot(uniform, dist='uniform', plot=pylab)
pylab.show()
```

```python
stats.skew(uniform)
```
