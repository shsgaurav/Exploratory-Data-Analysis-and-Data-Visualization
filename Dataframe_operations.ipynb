```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
dataframe = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
target = pd.DataFrame(data.target)
```

```python
dataframe['target'] = target
```

```python
dataframe.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dataframe.shape[0]
```

```python
dataframe['target'].value_counts()
```

```python
dataframe.plot(kind='scatter', x='sepal length (cm)', y= 'sepal width (cm)')
plt.grid()
```

```python
dataframe.plot(kind='scatter', x='petal length (cm)', y= 'petal width (cm)')
plt.grid()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(dataframe, hue='target', size=4) \
   .map(plt.scatter, 'petal length (cm)','petal width (cm)') \
   .add_legend();
```

```python
sns.pairplot(dataframe, hue='target', size=3)
```

```python
sns.FacetGrid(dataframe, hue='target', size=5) \
.map(sns.distplot, 'petal length (cm)') \
.add_legend()
```

```python
counts, bin_edges = np.histogram(dataframe['petal length (cm)'], bins=10, density=True)
pdf = counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```
